```python
# Bar Plot: Liver Cancer Cases by Country
plt.figure(figsize=(12, 5))
df["Country"].value_counts().head(10).plot(kind="bar", color="skyblue")
plt.title("Top 10 Countries with Most Liver Cancer Cases")
plt.xlabel("Country")
plt.ylabel("Number of Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# Gender Distribution
plt.figure(figsize=(6, 4))
df["Gender"].value_counts().plot(kind="bar", color=["blue", "pink"])
plt.title("Liver Cancer Cases by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(8, 4))
sns.histplot(df["Age"], bins=30, kde=True, color="purple")
plt.title("Age Distribution of Liver Cancer Cases")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Impact of Alcohol Consumption
plt.figure(figsize=(8, 4))
df.groupby("Alcohol_Consumption")["Prediction"].value_counts().unstack().plot(kind="bar", stacked=True)
plt.title("Liver Cancer Cases by Alcohol Consumption Level")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Count")
plt.show()
```

```python
# Rural vs. Urban Cases
plt.figure(figsize=(6, 4))
df.groupby("Rural_or_Urban")["Prediction"].value_counts().unstack().plot(kind="bar", stacked=True)
plt.title("Liver Cancer Cases in Rural vs. Urban Areas")
plt.xlabel("Living Area")
plt.ylabel("Count")
plt.show()
```

```python
# Count occurrences of each region
region_counts = df["Region"].value_counts()

# Plot using Seaborn for a better look
plt.figure(figsize=(12, 6))
sns.barplot(x=region_counts.index, y=region_counts.values, palette="magma",hue=region_counts.index , legend = False)

# Labels and title
plt.xlabel("Region")
plt.ylabel("Count")
plt.title("Distribution of Data by Region")
plt.xticks(rotation=45)  # Rotate labels for better visibility

# Show the plot
plt.show()
```

```python
# Calculate percentage of liver cancer cases by Gender
gender_pred = df[df["Prediction"] == "Yes"]["Gender"].value_counts(normalize=True) * 100  

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=gender_pred.index, y=gender_pred.values, palette="coolwarm", hue=gender_pred.value, legend = False)

# Labels and title
plt.xlabel("Gender")
plt.ylabel("Percentage (%)")
plt.title("Liver Cancer Cases by Gender (%)")

# Show values on bars
for i, value in enumerate(gender_pred.values):
    plt.text(i, value + 1, f"{value:.1f}%", ha="center", fontsize=12)

plt.ylim(0, 100)
plt.show()
```
